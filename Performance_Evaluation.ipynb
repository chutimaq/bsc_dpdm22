```python
from sklearn import datasets

iris_data = datasets.load_iris()
```

```python
X = iris_data['data']

y = iris_data['target']
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, KFold
```

```python
model = DecisionTreeClassifier()

# Define the k-fold cross validation
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# Train and evaluate the model using 10-fold cross validation
scores = cross_val_score(model, X, y, cv=kf)

# Print the average accuracy and standard deviation across all folds
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
model = DecisionTreeClassifier(max_depth=2)

# Define the k-fold cross validation
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# Train and evaluate the model using 10-fold cross validation
scores = cross_val_score(model, X, y, cv=kf)

# Print the average accuracy and standard deviation across all folds
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
from sklearn.model_selection import LeaveOneOut
```

```python
model = DecisionTreeClassifier(max_depth=2)

# Define the k-fold cross validation
LOO = LeaveOneOut()

# Train and evaluate the model using 10-fold cross validation
scores = cross_val_score(model, X, y, cv=LOO)

# Print the average accuracy and standard deviation across all folds
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

# HW6 วนลูปทำ 10-fold cross validation

```python
from sklearn.model_selection import KFold
from sklearn.tree import DecisionTreeClassifier

# define number of folds
k = 10

# create k-fold cross-validation
kf = KFold(n_splits=k)

# initialize the decision tree classifier
clf = DecisionTreeClassifier()

# loop through each fold
for train_index, test_index in kf.split(shuffled_df):
    
    # get the training and testing sets for this fold
    train_set = shuffled_df.iloc[train_index]
    test_set = shuffled_df.iloc[test_index]
    
    # train the classifier on the training set
    clf.fit(train_set[columns], train_set['label'])
    
    # evaluate the classifier on the testing set
    accuracy = clf.score(test_set[columns], test_set['label'])
    
    # print the accuracy for this fold
    print('Accuracy: {:.2f}%'.format(accuracy*100))
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression

num_folds = 10
folds = np.array_split(shuffled_df, num_folds)

for fold_idx in range(num_folds):
    # Get test set
    test_set = folds[fold_idx]
    
    # Get training set
    train_folds = folds[:fold_idx] + folds[fold_idx+1:]
    train_set = pd.concat(train_folds)
    
    # Train model
    model = LogisticRegression()
    model.fit(train_set[columns], train_set['label'])
    
    # Evaluate model on test set
    accuracy = model.score(test_set[columns], test_set['label'])
    
    print(f"Fold {fold_idx+1} accuracy: {accuracy:.3f}")
```
