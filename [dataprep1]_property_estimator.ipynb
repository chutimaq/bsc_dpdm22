##HW1 เปรียบเทียบความแตกต่างในการประเมินราคาทรัพย์แต่ละทรัพย์ของนักประเมินแต่ละคนและของราคาตั้งขาย ใช้ MAPE เป็นเกณฑ์ในการเปรียบเทียบ

###Home

```python
ass2['ประเมิน1'] = ass2['post id'].map(ass1.set_index('post id')['ประเมินราคาซื้อ-ขาย']);
ass2
```

```python
ass2['ประเมิน2'] = ass2['post id'].map(ass2.set_index('post id')['ประเมินราคาซื้อ-ขาย']);
ass2
```

```python
ht = ass2[['post id','ราคาทรัพย์','ประเมิน1','ประเมิน2']]
ht
```

```python
ht.isnull().any()
```

```python
ht = ht.fillna(value={'ประเมิน2':ass2['ประเมิน2'].mean()})
ht
```

```python
ht.isnull().any()
```

#### MAPE

```python
from sklearn.metrics import mean_absolute_percentage_error as mape
```

```python
ac = ht['ราคาทรัพย์']
pr1 = ht['ประเมิน1']
pr2 = ht['ประเมิน2']
```

```python
pred1 = mape(ac, pr1)
pred2 = mape(ac, pr2)
mape_h = mape(pr1,pr2)
```

```python
print(f'ค่า MAPE ของผู้ประเมินคนที่ 1 เท่ากับ {pred1:.3f}\nค่า MAPE ของผู้ประเมินคนที่ 2 เท่ากับ {pred2:.3f}\nค่า MAPE ราคาตั้งขายของผู้ประเมินคนที่ 1 และ ผู้ประเมินคนที่ 2  เท่ากับ {mape_h:.3f}')
```
